```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy.random as np
import sys
import matplotlib

%matplotlib inline
```

```python
np.seed(111)

# Function to generate test data
def CreateDataSet(Number=1):
    
    Output = []
    
    for i in range(Number):
        
        # Create a weekly (mondays) date range
        dod = pd.date_range(start='1/1/2009', end='12/31/2012', freq='W-MON')
        
        # Create random data
        data = np.randint(low=25,high=1000,size=len(dod))
        
        # Status pool
        status = ['In Custody', 'Not in Custody']
        
        # Make a random list of statuses
        random_status = [status[np.randint(low=0,high=len(status))] for i in range(len(dod))]
        
        # governate pool
        governate = ['Damascus','Rural Damascus','Qunietra','Aleppo','Homs','Idlib']
        
        # Make a random list of governates 
        random_governate = [governate[np.randint(low=0,high=len(governate))] for i in range(len(dod))]
    
        Output.extend(zip(random_governate, random_status, data, dod))
        
    return Output
```

```python
# Calculate Outliers
StateYearMonth = Daily.groupby([Daily.index.get_level_values(0), Daily.index.get_level_values(1).year, Daily.index.get_level_values(1).month])
Daily['Lower'] = StateYearMonth['Body Count'].transform( lambda x: x.quantile(q=.25) - (1.5*x.quantile(q=.75)-x.quantile(q=.25)) )
Daily['Upper'] = StateYearMonth['Body Count'].transform( lambda x: x.quantile(q=.75) + (1.5*x.quantile(q=.75)-x.quantile(q=.25)) )
Daily['Outlier'] = (Daily['Body Count'] < Daily['Lower']) | (Daily['Body Count'] > Daily['Upper']) 

# Remove Outliers
Daily = Daily[Daily['Outlier'] == False]
```

```python
Daily.head()
```

```python
# Combine all markets

# Get the max customer count by Date
ALL = pd.DataFrame(Daily['Body Count'].groupby(Daily.index.get_level_values(1)).sum())
ALL.columns = ['Body Count'] # rename column

# Group by Year and Month
YearMonth = ALL.groupby([lambda x: x.year, lambda x: x.month])

# What is the max customer count per Year and Month
ALL['Max'] = YearMonth['Body Count'].transform(lambda x: x.max())
ALL.head()
```

```python
data = [1000,2000,3000]
idx = pd.date_range(start='12/31/2011', end='12/31/2013', freq='A')
Estimate = pd.DataFrame(data, index=idx, columns=['Estimate'])
Estimate
```

```python
combined = pd.concat([ALL,Estimate], axis=0)
combined = combined.sort_index(axis=0)
combined.tail()
```

```python
fig, axes = plt.subplots(figsize=(12, 7))

combined['Estimate'].fillna(method='pad').plot(color='green', label='Estimate')
combined['Max'].plot(color='blue', label='All Markets')
plt.legend(loc='best');
```
